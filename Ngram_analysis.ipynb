```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from collections import Counter
import matplotlib.pyplot as plt

# Download required NLTK resources
nltk.download('punkt_tab')

def generate_ngrams(text, n):
    tokens = word_tokenize(text.lower())  # Tokenizing and converting to lowercase
    return list(ngrams(tokens, n))

def plot_ngram_frequencies(ngram_freq, title):
    ngrams, frequencies = zip(*ngram_freq.most_common(5))  # Top 5 most common n-grams
    ngrams = [" ".join(ng) for ng in ngrams]

    plt.figure(figsize=(8, 5))
    plt.bar(ngrams, frequencies, color='skyblue')
    plt.title(title)
    plt.xlabel("N-grams")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

# Input text
text = "Natural Language Processing is a field of AI. Language models are used in NLP. AI is advancing rapidly."

# Generate unigrams, bigrams, and trigrams
unigrams = generate_ngrams(text, 1)
bigrams = generate_ngrams(text, 2)
trigrams = generate_ngrams(text, 3)

# Frequency analysis
unigram_freq = Counter(unigrams)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

# Display most frequent N-grams
print("Top 5 Unigrams:", unigram_freq.most_common(5))
print("Top 5 Bigrams:", bigram_freq.most_common(5))
print("Top 5 Trigrams:", trigram_freq.most_common(5))

# Visualization
plot_ngram_frequencies(bigram_freq, "Top 5 Most Frequent Bigrams")
plot_ngram_frequencies(trigram_freq, "Top 5 Most Frequent Trigrams")
```
